```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('wine_data.csv',header = None)
```

```python
data
```

```python
data.columns = ['Id','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline ']
```

```python
data
```

```python
data.drop('Id',axis = 'columns')
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(data)
```

```python
wcss = []
for i in range(1,11):
  k_means = KMeans(n_clusters=i, init='k-means++')
  k_means.fit(x)  
  wcss.append(k_means.inertia_) 
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Value of k')
plt.ylabel('WCSS')
```

Inference: According to the elbow graph, the ideal number of clusters is 3

```python
k_means = KMeans(n_clusters = 3,init='k-means++', random_state= 0)
y_pred = k_means.fit_predict(x)
```

```python
plt.scatter(x[y_pred==0, 0], x[y_pred==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(x[y_pred == 1,0], x[y_pred == 1,1], s = 100, c = 'cyan', label = 'Cluster 2')
plt.scatter(x[y_pred == 2,0], x[y_pred == 2,1], s =100, c='green',label = 'Cluster 3')
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], s=100, c='yellow', label = 'Centroids')
plt.title('Clusters of Wines')

plt.show()
```
