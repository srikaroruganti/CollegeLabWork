```python
import pandas as pd
import seaborn as se
import matplotlib
import matplotlib.pyplot as plt
```

Loading the dataset into pandas dataframe

```python
data = pd.read_csv('train.csv')
```

Displaying the entries

```python
data
```

Checking the frequency of each passenger class

```python
data['Pclass'].value_counts()
```

Bar Graph for number of males, females and mean of fare

```python
res = data['Sex'].value_counts()

titles = ["Male", "Female", "Mean of Fare"]
vals = [res[0], res[1], data["Fare"].mean()]

plt.bar(titles, vals)
plt.show()
```

Line chart

```python
data.iloc[50:100].dropna().plot(kind = 'line', x = 'PassengerId',y = 'Fare')
plt.show()
```

Line chart for multiple values

```python
ax = plt.gca()
data.iloc[0:21].dropna().plot(kind = 'line',x = 'Name',y = 'Age',color = 'blue',ax = ax)
data.iloc[0:21].dropna().plot(kind = 'line',x = 'Name',y = 'Fare',color = 'Red',ax = ax)
plt.figure(figsize = (100,200))
plt.show()
```

Histogram


```python
plt.hist(data['Age'],bins = 8,color = 'green',edgecolor = 'black')
plt.show()
```

Boxplot

```python
bpl = data.boxplot(column=['Age','Fare','Pclass'])

bpl.plot()
plt.show()
```

Scatter Plot


```python
plt.scatter(data['Age'],data['Fare'],color='red')
plt.show()
```

Scatter plot 3D

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(data['Age'], data['Fare'], color = 'blue')

plt.show()
```
