```python
#g
BMI_zero = []
BMI_one = []
for i in range(len(data_frame['BMI'])):
    if data_frame['OUTCOME'][i] == 0:
        BMI_zero.append(data_frame['BMI'][i])
    else:
        BMI_one.append(data_frame['BMI'][i])
```

```python
def median(lst):
    sortedLst = sorted(lst)
    lstLen = len(lst)
    index = (lstLen - 1) // 2
    if (lstLen % 2):
        return sortedLst[index]
    else:
        return (sortedLst[index] + sortedLst[index + 1])/2.0
median_BMI_zero = median(BMI_zero)
median_BMI_one = median(BMI_one)
print("Median of BMI -- OUTCLASS '0': ",median_BMI_zero)
print("Median of BMI -- OUTCLASS '1': ",median_BMI_one)
```

```python
#i
data_frame.loc[(data_frame['AGE']==21),'AGE_CATEGORY'] = 'Young'
data_frame.loc[(data_frame['AGE']>21) & (data_frame['AGE']<50),'AGE_CATEGORY'] = 'Middle_Young'
data_frame.loc[(data_frame['AGE']>=50),'AGE_CATEGORY'] = 'Old'
```

```python
#j
cols = ["GLUCOSE", "BLOODPRESSURE", "SKINTHICKNESS", "INSULIN"]
data_frame[cols] = data_frame[cols].replace({0:np.nan})
data_frame.head()
```

```python
def replace_NAN(data_frame,column):
    mean = np.mean(data_frame[column])
    mean = round(mean,2)
    data_frame[column].fillna(value=mean, inplace=True)
```

```python
for i in cols:
    replace_NAN(data_frame,i)
```
